```python
import math
def distance(n,p1,p2):
    sum=0
    for i in range(n):
        sum+=math.pow(p1[i]-p2[i],2)
    return math.sqrt(sum)
def centroid(cluster):
    import operator
    
    sum=(0,0,0)
    for i in range(len(cluster)):
        sum=tuple(map(operator.add, sum, data[cluster[i]-1]))
    sum=list(sum)
    for i in range(len(sum)):
        sum[i]/=len(cluster)
    
    return tuple(sum)
```

```python
k=2
from random import randint
c=[]
c_new=[]
clusters=[]
for i in range(k):
    c.append(None)
    c_new.append(None)
    c[i]=data[randint(0, len(data)-1)]
    clusters.append([])
print(c)
iteration=1
while(True):
    print("\n",iteration,":")
    for i in range(len(data)):
        pos=None
        min_dist=9999
        for j in range(k):
            dist=distance(3,c[j],data[i])
            if dist<min_dist:
                min_dist=dist
                pos=j
        clusters[pos].append(i+1)
    flag=0
    for i in range(k):
        c_new[i]=centroid(clusters[i])
        print(c_new[i],c[i])
        if( all(j == k for j, k in zip(c_new[i],c[i])))== False:
            flag=1
    print (clusters)
    if flag==0:
        break
    else:
        for i in range(k):
            clusters[i]=[]
            c[i]=tuple(list[c_new[i]])
    
    iteration+=1
    if iteration==10:
        break
    
```
